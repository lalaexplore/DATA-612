Based on the articles provided, there seem to be several ways to mitigate the harm caused by recommender systems. One idea that stood out to me is the importance of the data being fed into these systems. This is crucial because, even before machines existed, humans tended to make choices based on what was popular or what generated the most profit. However, the real issue lies in the quality of those choices and how they may affect people over time.
The same concept applies to recommender systems. When systems are trained on content that is popular but harmful, they are more likely to recommend that content to users. I believe that the data collected for these systems should not be the sole responsibility of data scientists; it requires the efforts of an entire team. For example, cybersecurity professionals can help screen users, authors, and content creators to ensure that harmful or misleading content is identified.
Ordinary people also play an important role in this discussion. We should ask questions such as: Are the things we are feeding into these systems humane, ethical, and unbiased? Are we making the right decisions? Ultimately, I believe that mitigating the harm of recommender systems comes down to human judgment, critical thinking, and collective responsibility. It will take the combined efforts of many people to create recommender systems that are more ethical, fair, and beneficial to society.
