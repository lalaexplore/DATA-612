*After reviewing the recommendation system used by Metacritic, Rotten Tomatoes, and IMDb, I believe Rotten Tomatoes had an advantage over the other two, even though IMDb and Metacritic are owned by the same company. Unlike personalized recommender systems that use content-based filtering or collaborative filtering, Rotten Tomatoes uses a non-personalized aggregation technique, It collects reviews from professional critics and audience members and summarizes them into two separate scores: the Tomatometer score (professional critics) and the Popcorn meter score (audience ratings). *Meta, the parent company of Instagram, has taken steps to address this issue by monitoring accounts that repeatedly repost unoriginal content and limiting their reach. This helps ensure that original creators receive proper credit and have a better chance of being discovered by new audiences.*

*Similar to the IMDb example, groups of users can influence recommendation systems in ways that are unfair to creators.
