The most interesting point for me in Christopher Johnson’s talk was the comparison between the full gridify method and the half gridify method. From my understanding, the full gridify method stores subsets of rating data for an item or user across multiple partitions. While this approach requires less local memory, it also involves a significant amount of shuffling and aggregation of intermediate data to group users and calculate the necessary sums. In contrast, the half gridify method stores the history of a vector (either a user or an item) within separate partitions, where it remains in memory. Although this requires more local memory and may require copies of item vectors that may not fit into memory, it offers several advantages. The ALS running times are faster than those of the Hadoop and Spark full gridify methods because the ratings data is already stored in each partition. As a result, there is no unnecessary data shuffling during each iteration and no need to send full copies of vectors to all worker partitions.
