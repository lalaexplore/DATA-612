Recommender systems can reinforce human bias because they learn from user behaviour and historical data. If users have historically favored certain groups, products, or viewpoints, the system may continue recommending those same options, creating a feedback loop that strengthens existing biases. For example, a job recommendation platform might show technical positions more frequently to men if historical data reflects gender imbalances in those fields.

I found an article from UN Women on how AI reinforces gender bias and how it learns from the historical data we feed into it. The article discusses the long-term effects of biased training data and highlights the importance of actively selecting datasets that reflect diverse social backgrounds, cultures, and roles while removing historical biases, such as associating certain jobs or traits with a specific gender. Del Villar also suggested that development teams should be diverse in terms of race, gender, and cultural background so that different perspectives can help reduce blind spots that may lead to biased AI systems. She further emphasized the importance of awareness and education because biased AI can influence decision making processes in many areas.

As for preventing unethical targeting or customer segmentation, I believe recommender systems can help reduce these issues when they are designed responsibly. For instance, two presentations from the RecSys Conference demonstrated different approaches to recommender systems. One presentation focused on safety-aware recommendations, while the other explored recommending high-risk, high-reward items to create a better balance between popular and less popular recommendations. These studies showed that recommender systems can be designed to mitigate unethical targeting and create fairer customer segmentation strategies.


Although recommender systems can be designed to reduce bias, they can also reinforce it when developers prioritize engagement, clicks, or profit without considering fairness. As a result, the impact of a recommender system depends largely on the quality of the data and the design choices made during development.
