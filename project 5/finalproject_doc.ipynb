# **Final Project Proposal** by Sandra Dela Cruz

For my final project, I will continue using the MovieLens 1M dataset from Project 5. In the previous project, I implemented a baseline recommendation system using the Alternating Least Squares (ALS) algorithm in Apache Spark. This model generated recommendations based on user–item interactions and was evaluated using RMSE, MAE, Precision@K, and Recall@K.

For the final project, I will extend this system by building a hybrid recommender that combines collaborative filtering with content-based information. In addition to user ratings, I will incorporate movie metadata using the directedBy feature.

The system will first use the ALS model to generate baseline recommendations. It will then enhance these recommendations by using director information to adjust rankings, giving higher priority to movies that share similar directors with films a user has previously liked. This allows the system to incorporate both user behavior patterns and content-level similarities.

The hybrid model will be evaluated using the same metrics as the baseline system (RMSE, MAE, Precision@K, and Recall@K). In addition, I will include diversity and novelty metrics to evaluate the variety and usefulness of recommendations beyond accuracy alone.

Finally, I will compare the baseline ALS model with the hybrid model to determine whether incorporating metadata improves recommendation quality, both in terms of prediction accuracy and recommendation diversity.
